## Bivariate and Multivariate analysis

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

### Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(tips['total_bill'],tips['tip'], hue=tips['sex'], style=tips['smoker'],size=tips['size'])
```

### Bar Plot (Numerical - Categorical)

```python
sns.barplot(titanic['Pclass'], titanic['Age'], hue=titanic['Gender'])
```

```python
sns.barplot(titanic['Pclass'], titanic['Fare'])
```

### Box Plot (Numerical - Catagorical)

```python
sns.boxplot(titanic['Gender'], titanic['Age'], hue=titanic['Survived'])
```

### Displot (Numerical - Categorical)

```python
sns.displot(titanic[titanic['Survived']==0]['Age'])
```

### HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']), annot=True)
```

```python
titanic.groupby('Embarked').mean()['Survived']*100
```

### ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

### PairPlot

```python
sns.pairplot(iris, hue='species')
```

### LinePlot (Numerical - Numerical) use on time based features which is incresing by time

```python
new = flights.groupby('year').sum().reset_index()
new
```

```python
sns.lineplot(new['year'], new['passengers'])
```

```python
flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
